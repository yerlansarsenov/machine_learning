```python
import numpy as np
from scipy import spatial as sp

def cosdist(a, b):
    dott = np.dot(a, b)
    magn1 = np.linalg.norm(a)
    magn2 = np.linalg.norm(b)
    return 1 - dott / (magn1 * magn2)

def normalize(word_vec):
    norm=np.linalg.norm(word_vec)
    if norm == 0: 
       return word_vec
    return word_vec/norm

a = normalize((-2,3))
b = normalize((4,3))

c = normalize((-6,1))
c1 = normalize((6, -1))

print(cosdist(a, c1))

print(sp.distance.cosine(a, c1))
```
